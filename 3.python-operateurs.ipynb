# Opérateurs

### Les opérateur de comparaison

Nous les avons déjà vu auparavant:
  - **`==`** (equal to)
  - **`!=`** (not equal to)
  - **`<`** (less than)
  - **`<=`** (less than or equal to)
  - **`>`** (greater than)
  - **`>=`** (greater than or equal to)

**Les autres opérateurs sont:**
- opérateurs arithmétiques
  - **`+`** (addition)
  - **`-`** (subtraction)
  - **`*`** (multiplication)
  - **`/`** (division)
  - __`**`__ (exponent)

```python
print(1+2)
print(2-1)
print(2*1)
print(2/2)
print(2**4)
```

```python
print(type(1))
```

```python
print(type(1.0))
```

- opérateur d'attribution
  - **`=`** (assign a value)
  - **`+=`** (add and re-assign; increment)
  - **`-=`** (subtract and re-assign; decrement)
  - **`*=`** (multiply and re-assign)


```python
i =  1
i += 1 # i = i + 1
i -= 1 # i = i - 1
i *= 1 # i = i * 1
i /= 1 
```
